## 解析 json

```python
import requests
import json

response = requests.get('https://httpbin.org/get')
print(type(response.text))

#以下兩個是一樣的
print(response.json())
print(json.loads(response.text))

print(type(response.json))
```

## 獲取二進位數據

```python
import requests

response = requests.get('http://github.com/favicon.ico')
print(type(response.text), type(response.content))

#str
print(response.text)

#bytes 二進制編碼
print(response.content) 
```

## 寫入二進位數據，儲存成檔案

```python
import requests

response = requests.get('http://github.com/favicon.ico')

#寫入檔案，以二進為形式
with open('favicon.ico','wb') as f:
    f.write(response.content)
    f.close()
```

## 基本POST請求

```python
import requests

data = {
    'name':'germany',
    'age':22
}

#直接傳送一個字典
response = requests.post("http://httpbin.org/post", data=data)
print(response.text)
```

```python
import requests

data = {
    'name':'germany',
    'age':22
}

headers = {
     "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36"
}

#直接傳送一個字典
response = requests.post("http://httpbin.org/post", data=data, headers=headers)
print(response.json())
```

## status_code 判斷

```python
import requests

response = requests.get("http://www.jianshu.com")

if not response.status_code == requests.codes.ok :
    exit() 
else: 
    print('Request successfully')
```

## 文件上傳: files參數

## 獲取cookies

```python
import requests

response = requests.get('http://www.baidu.com')
print(type(response.cookies))
print(response.cookies)

for key,value in response.cookies.items():
    print(key+" = "+value)
```

## 會話維持: requests.Session()

模擬登入

```python
import requests

#獨立瀏覽器 1
requests.get('https://httpbin.org/cookies/set/number/1234')

#獨立瀏覽器 2
response = requests.get('https://httpbin.org/cookies')

#得不到我們要的cookies
print(response.text)
```

```python
import requests

#想像成把兩個事情在同一個瀏覽器完成
s = requests.Session()
s.get('https://httpbin.org/cookies/set/number/1234')
response = s.get('https://httpbin.org/cookies')

print(response.text)
```

## 代理設置 VPN

```python
import requests

proxies = {
    'http':'http://127.0.0.1:9743'
    'https':'https://127.0.0.1:9743'
}

response = requests.get('http://www.baidu.com', proxies = proxies)
print(response.status_code)
```

```python
import requests

#proxy有密碼
proxies = {
    'http':'http://user:password@127.0.0.1:9743'
}

response = requests.get('http://www.baidu.com', proxies = proxies)
print(response.status_code)
```

## 設置超時

```python
import requests
from requests.exceptions import ReadTimeout

try:
    response = requests.get('https://www.baidu.com', timeout = 0.1)
    print(reponse.status_code)

except ReadTimeout:
    print('Time out!!')
```

## 認證設置 

```python
import requests
from requests.auth import HTTPBasicAuth

#沒有帳號密碼會被禁止請求
r = requests.get('http://120.27.34.24:9001')
print(r.status_code)
```

```python
import requests
from requests.auth import HTTPBasicAuth

r = requests.get('http://120.27.34.24:9001', auth=HTTPBasicAuth('user','123'))
print(r.status_code)
```

```python
import requests
from requests.auth import HTTPBasicAuth

r = requests.get('http://120.27.34.24:9001', auth=('user','123'))
print(r.status_code)
```

## 異常處理，可以保證程式不要被中斷的運行

```python
import requests
from requests.exceptions import ReadTimeout, HTTPError, RequestException

try:
    response = requests.get('https://www.baidu.com', timeout = 0.1)
    print(response.status_code)

except ReadTimeout:
    print('Timeout')
except HTTPError:
    print('Http error')
except RequestException:
    print('Request error')
```
